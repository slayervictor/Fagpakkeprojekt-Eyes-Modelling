```python
from textwrap import wrap
import glob
import os
def process_txt_to_txt_file(txt_path,output_folder):
    with open(txt_path,'r') as file:
        content = file.read()
    
    
    # Set the same format
    content= content.replace("Fill-in-the-blanks Questions:","Fill-in-the-blank Questions:")
    content= content.replace("Fill-in-the-Blanks Questions:","Fill-in-the-blank Questions:")


    mcq_identifier = "Multiple Choice Questions:"
    fib_identifier = "Fill-in-the-blank Questions:"
    

    mcq_start = content.find(mcq_identifier) + len(mcq_identifier)
    fib_start = content.find(fib_identifier) + len(fib_identifier)

    main_text = content[:content.find(mcq_identifier)].strip()
    mcq_text = content[mcq_start:content.find(fib_identifier)].strip()
    fib_text = content[fib_start:].strip()

    # set the same format 
    mcq_text = mcq_text.replace(")",".")
    mcq_text = mcq_text.replace("?","?\n")
    mcq_text = mcq_text.replace("c.","\nc.")
    mcq_text = mcq_text.replace(" a.","a.")
    mcq_text = mcq_text.replace("A.","a.")
    mcq_text = mcq_text.replace("\nB.","b.")
    mcq_text = mcq_text.replace("C.","c.")
    mcq_text = mcq_text.replace("\nD","d.")

    main_text = main_text +"."
    main_text = main_text.replace("..",".")

    formatted_text = "Text\n\n" + main_text
    formatted_mcq = "Multiple Choice Questions:\n\n" + mcq_text
    formatted_fibq = "Fill-in-the-blank Questions:\n\n" + fib_text

    base_filename = os.path.splitext(os.path.basename(txt_path))[0]
    text_filename = os.path.join(output_folder, f"{base_filename}_text.txt")
    mcq_filename = os.path.join(output_folder, f"{base_filename}_MCQ.txt")
    fib_filename = os.path.join(output_folder, f"{base_filename}_FIBQ.txt")

    with open(text_filename, 'w', encoding='utf-8') as f:
        f.write(formatted_text)
    with open(mcq_filename, 'w', encoding='utf-8') as f:
        f.write(formatted_mcq)
    with open(fib_filename, 'w', encoding='utf-8') as f:
        f.write(formatted_fibq)

    print(f"Processed file: {txt_path}")
# Folder paths (Updated as per your provided paths)
input_folder = 'C:\\Users\\s224228\\Documents\\Fagpakkeprojekt-Eyes-Modelling\\Manglende passages'
output_folder = 'C:\\Users\\s224228\\Documents\\Fagpakkeprojekt-Eyes-Modelling\\eye_tracking_expiriment\\files\\Files_txt'

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

for txt_path in glob.glob(os.path.join(input_folder, '*.txt')):
    process_txt_to_txt_file(txt_path, output_folder)
```
