```python
from striprtf.striprtf import rtf_to_text
from textwrap import wrap
import glob
import os

def convert_and_wrap_rtf_to_txt(rtf_path, output_folder, wrap_width=80):
    with open(rtf_path, 'r') as file:
        rtf_content = file.read()

    # Convert RTF content to text
    text = rtf_to_text(rtf_content)

    # Wrap text uniformly
    wrapped_text = []
    paragraphs = text.split('\n\n')  # Assuming paragraphs are separated by two newlines
    for para in paragraphs:
        wrapped_paragraph = '\n'.join(wrap(para, width=wrap_width))
        wrapped_text.append(wrapped_paragraph)
    uniform_text = '\n\n'.join(wrapped_text)

    # Define the output file path
    base_filename = os.path.splitext(os.path.basename(rtf_path))[0]
    txt_filename = os.path.join(output_folder, base_filename + ".txt")

    # Save the uniformly wrapped text to a new TXT file
    with open(txt_filename, 'w', encoding='utf-8') as txt_file:
        txt_file.write(uniform_text)

    print(f"Converted and wrapped {rtf_path} to {txt_filename}")

# Folder paths (Updated as per your provided paths)
input_folder = 'C:\\Users\\s224228\\Documents\\Fagpakkeprojekt-Eyes-Modelling\\Ting fra supervisor\\Text_passges'
output_folder = 'C:\\Users\\s224228\\Documents\\Fagpakkeprojekt-Eyes-Modelling\\files'
# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Convert and wrap all RTF files in the input folder
for rtf_filename in glob.glob(os.path.join(input_folder, '*.rtf')):
    convert_and_wrap_rtf_to_txt(rtf_filename, output_folder)
```
