```python
import tkinter as tk

def read_text(filename):
    with open(filename, 'r') as file:
        return file.read()

Sequence = ['start.txt',
'Ai_HC_P01_text.txt','Ai_HC_P01_MCQ.txt','Ai_HC_P01_FIBQ.txt',
'Ai_HC_P02_text.txt','Ai_HC_P02_MCQ.txt','Ai_HC_P02_FIBQ.txt',
'Ai_HC_P03_text.txt','Ai_HC_P03_MCQ.txt','Ai_HC_P03_FIBQ.txt',
'Ai_HC_P04_text.txt','Ai_HC_P04_MCQ.txt','Ai_HC_P04_FIBQ.txt',
'Ai_HC_P05_text.txt','Ai_HC_P05_MCQ.txt','Ai_HC_P05_FIBQ.txt',
'Ai_HC_P06_text.txt','Ai_HC_P06_MCQ.txt','Ai_HC_P06_FIBQ.txt',
'slut.txt']
texts=[]

for filename in Sequence:
    texts.append(read_text(filename))

# Setting the font and size
font_size=[20]
font_family=['Arial']

current_text_index = 0


root = tk.Tk()
root.title('Text display')
#root.geometry('1600x1200')
root.attributes('-fullscreen', True)
root.resizable(width=False, height=False)

def close_window(event=None):
    root.destroy()

# Create the text widget

canvas = tk.Canvas(root, width=1600, height=1200)
canvas.pack()


# Start position for the text

x_position = 50
y_position = 100

def draw_text():
    global current_text_index
    canvas.delete('all')
    canvas.create_text(x_position, y_position, text=texts[current_text_index], font=(font_family[0], font_size[0]), anchor='nw')
    canvas.update()


draw_text()

def navigate_text(event):
    global current_text_index
    if event.keysym == 'Right':
        if current_text_index < len(texts) - 1:
            current_text_index += 1
            draw_text()
    elif event.keysym == 'Left':
        if current_text_index > 1:
            current_text_index -= 1
            draw_text()


# Bind arrow keys to the navigate_text function
root.bind('<Left>', navigate_text)
root.bind('<Right>', navigate_text)

root.bind('<Escape>', close_window)

# Start the GUI event loop
root.mainloop()
```
