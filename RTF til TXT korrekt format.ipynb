```python
import os
import re
import glob


def format_mcq(mcq_text):
    formatted_text = "Multiple Choice Questions:\n"
    lines = (mcq_text.replace('\n',' ').replace('?','?\n')).split('\n')

    # Strip leading spaces from each line
    stripped_lines = [line.strip() for line in lines]

    # Join the lines back together into a single string
    formatted_mcq_text = '\n'.join(stripped_lines)
    formatted_mcq_text = formatted_mcq_text.replace('c.','\nc.')
    lines = formatted_mcq_text.split('\n')
    for line in lines:
        is_d = line.find('d.')
        if is_d!=-1:
            where_capital=[letter for letter in range(len(line)) if line[letter].isupper()]
            if len(where_capital) <3:
                continue
            index=where_capital[2]-1
            s=list(line)
            s[index] = '\n'
            final_line="".join(s)
            line_index= lines.index(line)
            lines[line_index]=final_line
        formatted_mcq_text ="\n".join(lines)
    #formatted_mcq_text =
    #formatted_mcq_text =
    #formatted_mcq_text =
    #formatted_mcq_text =
    final_text=formatted_text + formatted_mcq_text

    return final_text

def format_fibq(fibq_text):
    formatted_text = formatted_text = "Fill-in-the-blank Questions:\n"
    #We remove all lines. Then add lines at the specific places we want. 
    fib_text_formatted = '\n'.join(line.strip() for line in fibq_text.replace('\n',' ').replace('.','.\n').split('\n'))

    return formatted_text + fib_text_formatted



def process_text_file(txt_path, output_folder):
    with open(txt_path, 'r', encoding='utf-8') as file:
        content = file.read()

    mcq_identifier = "Multiple Choice Questions:"
    fib_identifier = "Fill-in-the-blank Questions:"

    mcq_start = content.find(mcq_identifier) + len(mcq_identifier)
    fib_start = content.find(fib_identifier) + len(fib_identifier)

    main_text = content[:content.find(mcq_identifier)].strip()
    mcq_text = content[mcq_start:content.find(fib_identifier)].strip()
    fib_text = content[fib_start:].strip()

    formatted_mcq = format_mcq(mcq_text)
    formatted_fibq = format_fibq(fib_text)

    base_filename = os.path.splitext(os.path.basename(txt_path))[0]
    text_filename = os.path.join(output_folder, f"{base_filename}_text.txt")
    mcq_filename = os.path.join(output_folder, f"{base_filename}_MCQ.txt")
    fib_filename = os.path.join(output_folder, f"{base_filename}_FIBQ.txt")

    with open(text_filename, 'w', encoding='utf-8') as f:
        f.write(main_text)
    with open(mcq_filename, 'w', encoding='utf-8') as f:
        f.write(formatted_mcq)
    with open(fib_filename, 'w', encoding='utf-8') as f:
        f.write(formatted_fibq)

    print(f"Processed file: {txt_path}")
# Folder paths (Updated as per your provided paths)
input_folder = 'C:\\Users\\s224228\\Documents\\Fagpakkeprojekt-Eyes-Modelling\\files'
output_folder = 'C:\\Users\\s224228\\Documents\\Fagpakkeprojekt-Eyes-Modelling\\files\\Files_txt'

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

for txt_path in glob.glob(os.path.join(input_folder, '*.txt')):
    process_text_file(txt_path, output_folder)
```
