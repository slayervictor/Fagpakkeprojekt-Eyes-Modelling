```python
import tkinter as tk

# Load texts from a file
with open('texts.txt', 'r') as file:
    texts = [line.strip() for line in file.readlines()]

# Assuming these are your desired font sizes and families
font_sizes = [24, 30, 36, 42, 48]
font_families = ["Arial", "Verdana", "Times New Roman", "Helvetica", "Courier"]

# Current indices for text, font size, and font family
current_text_index = 0
current_size_index = 0
current_family_index = 0

# Create the main window
root = tk.Tk()
root.title("Text Display")
root.geometry("800x600")
root.resizable(False, False)

# Create a Canvas widget
canvas = tk.Canvas(root, width=800, height=600)
canvas.pack()

# Position for the text
x_position = 50
y_position = 100

# Global variables to hold current text details
current_text_details = {
    'text_content': '',
    'font_size': '',
    'font_name': ''
}

def draw_text():
    global current_text_index, current_size_index, current_family_index
    canvas.delete("all")  # Clear previous text
    text = texts[current_text_index]
    size = font_sizes[current_size_index]
    family = font_families[current_family_index]
    # Extract first word
    first_word = text.split(' ')[0] if ' ' in text else text
    canvas.create_text(x_position, y_position, text=first_word, anchor="nw", font=(family, size))

     # Update the global variable with current text details
    current_text_details = {
        'text_content': first_word,
        'font_size': size,
        'font_name': family
    }
    print(text)
    print(size)
    print(family)

draw_text()  # Initial drawing

def navigate_text(event):
    global current_text_index
    if event.keysym == 'Right':
        current_text_index = (current_text_index + 1) % len(texts)
    elif event.keysym == 'Left':
        current_text_index = (current_text_index - 1) % len(texts)
    draw_text()

# Bind arrow keys to the navigate_text function
root.bind('<Left>', navigate_text)
root.bind('<Right>', navigate_text)

# Start the GUI event loop
root.mainloop()

```
