```python
# data and code from https://github.com/axelroques/Eye_Movement_Classification/tree/master

from idt_algo import IDT
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gaze = pd.read_csv('test_case/gaze_data.csv')
eye = pd.read_csv('test_case/eye_data.csv')
eye.head(3)
```

```python
f, axes = plt.subplots(2, 1, figsize=(15, 8))

axes[0].plot(eye['t'], eye['x_right'], c='mediumblue', alpha=0.8, label='Right Eye')
axes[0].plot(eye['t'], eye['x_left'], c='crimson', alpha=0.8, label='Left Eye')
axes[0].set_ylabel('x', fontsize=15)


axes[1].plot(eye['t'], eye['y_right'], c='mediumblue', alpha=0.7, label='Right Eye')
axes[1].plot(eye['t'], eye['y_left'], c='crimson', alpha=0.7, label='Left Eye')
axes[1].set_ylabel('y', fontsize=15)
    
for ax in axes:
    ax.set_xlabel('Time', fontsize=15)
    ax.legend()
plt.tight_layout()
plt.show()
```

```python
t = eye['t']
x = eye['x_right']
y = eye['y_right']

classifier = IDT(t, x, y, threshold=2, window_size=20)
classifier.process()
classifier.plot()
```
